```python
is_started = False

print("""start - to start the car
stop  - to stop the car
quit  - to exit
""")

while True:
    inp = input(">> ").lower().strip()

    if inp == "start":
        if is_started:
            print("Car is already started!")
        else:
            is_started = True
            print("Car started...Ready to go!")

    elif inp == "stop":
        if not is_started:
            print("Car is already stopped!")
        else:
            is_started = False
            print("Car stopped.")

    elif inp == "help":
        print("""start - to start the car
stop  - to stop the car
quit  - to exit""")

    elif inp == "quit":
        print("Exited")
        break

    else:
        print("I don't understand that...")
```
