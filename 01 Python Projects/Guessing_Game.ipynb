```python
# user should have 5 guesses
# range of numbers from 1 to 100

import random
secret_number = random.randint(1, 100)

guesses = 5
print("Guess numbers between (1 to 100)")
while guesses != 0:
    guess = int(input(f"Guess: {abs(guesses-6)}"))
    if guess == secret_number:
        print("Hey!!!! You got it!")
        break;
    elif guess > secret_number:
        print("Too High")
    elif guess < secret_number:
        print("Too Low")
    else:
        pass  
    guesses -= 1
print(f"The secret number is {secret_number}")
```
